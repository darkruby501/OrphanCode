```python
posts2 = posts.groupby(['year','quarter']).agg({'likesCount' : np.sum,
                                              'commentsCount' : np.sum,
                                              'sharesCount' : np.sum,
                                              'id': 'count'})
posts2.rename(columns={'id':'postsCount'},inplace=True)
posts2['date_week'] = [datetime.datetime.strptime(str(a[0])+'-'+str(a[1]) + '-0', "%Y-%m-%w") for a in list(posts2.index)]
posts2.sort_index()
```

```python
ts = (a - np.datetime64('1970-01-01T00:00:00Z')) / np.timedelta64(1, 's')
ts
```

```python
posts2['ts'] = posts2['date_week'].apply(lambda x: (np.datetime64(x)-np.datetime64('1970-01-01T00:00:00Z')) / np.timedelta64(1, 's'))                                      
```
