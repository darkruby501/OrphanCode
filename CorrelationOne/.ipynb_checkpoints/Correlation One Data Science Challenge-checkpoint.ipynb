### Step 0: Load and Check Data

```python
## Load SQLite Database
conn = sqlite3.connect('./data/contentDiscovery.db')
```

```python
email_content = pd.read_sql_query('SELECT * FROM email_content', conn, parse_dates=['send_time'])
#email_content['send_time'] = pd.to_datetime(email_content['send_time'])
```

```python
email_content.tail()
```

```python
print('Number of article links sent out: {:,}'.format(email_content.shape[0]))
```

```python
## Load Apache Log Data
with open('./data/access.log') as f:
    lines = f.readlines()
```

```python
def log_parser(line):
    """Not proud of this function, but it's faster than mucking around with regexp or a log parser package"""
    line_data = {
        'date_clicked' : line[1:21],
        'article_id' : int(line[line.find('article_id=')+len('article_id='):line.find('&user')]),
        'user_id' : int(line[line.find('user_id=')+len('user_id='):line.find(' HTTP')]),
        'status_code' : int(line[line.find('HTTP/1.1" ')+len('HTTP/1.1" '):line.find('HTTP/1.1" ')+len('HTTP/1.1" ')+3])
    }
    return line_data

lines_parsed = []
for line in lines:
    lines_parsed.append(log_parser(line))
```

```python
clicks = pd.DataFrame(lines_parsed,columns=['date_clicked','user_id','article_id','status_code'])
clicks['date_clicked'] = pd.to_datetime(clicks['date_clicked'],format='%d/%b/%Y:%X')
```

```python
# Check for bad response
clicks.groupby('status_code').count()
```

#### Load Other Tables

```python
users = pd.read_sql_query('SELECT * FROM users', conn)
types = pd.read_sql_query('SELECT * FROM types', conn)
topics = pd.read_sql_query('SELECT * FROM topics', conn)

print('Number of Users = ', users.shape[0])
print('Number of Content Types = ', types.shape[0])
print('Number of Topics = ', topics.shape[0])
```

```python
users.head()
```

```python
articles = pd.read_sql_query("""SELECT article_id, topics.name AS topic, types.name AS type,
articles.topic_id AS "topic_id", articles.type_id as "type_id"
FROM articles 
LEFT JOIN topics ON articles.topic_id = topics.topic_id 
LEFT JOIN types ON articles.type_id = types.type_id"""
                          , conn)
```

```python
articles.head()
```

## Step 1: Merge DataFrames

```python
## Merge Apache Log with Emails
ctr = email_content.merge(clicks,on = ['user_id','article_id'],how='left')
ctr = ctr.merge(articles,on='article_id',how='inner')
ctr['clicked']=ctr['status_code'].notnull()
ctr['not_clicked']=ctr['status_code'].isnull()
```

```python
ctr.head()
```

## Step 2: Examine Overall Trend

```python
# Overall clickrate is 16.4%, that's pretty good, I'd say.

global_CTR = sum(ctr['clicked'])/ctr.shape[0]
print('Overall CTR: {0:.1f}%'.format(global_CTR*100))
```

```python
## Group article links by week
ctr_weekly = ctr.set_index(ctr['send_time']).resample('W',how='sum')
ctr_weekly['CTR'] = ctr_weekly['clicked']/(ctr_weekly['clicked']+ctr_weekly['not_clicked'])
```

```python
## Plot with rolling mean
pd.rolling_mean(ctr_weekly['CTR']*100,2).plot()
plt.ylabel('Click-Through Rate  (%)')
plt.xlabel('Date Sent')
plt.title('Global Click-Through Rate for Links')
```

```python
# See the raw numbers
np.round(ctr_weekly['CTR']*100,1)
```

```python
ctr_weekly[['clicked','not_clicked']].sum(axis=1).mean()*0.01
```

A change of 1% in click-through rate amounts to a difference of ~7500 more link viewings per week. How valuable is that?

## Variation in Proportion of Top Topics and Top Content Types Over Time
Perhaps global changes in the content delivered is responsible for the minor fluctations in CTR over time.

```python
topic_prop = ctr[ctr['topic'].isin(top_topics[0:10])].set_index('send_time').groupby([lambda x: x.weekofyear,'topic']).count()['content_id'].unstack()
topic_prop.divide(topic_prop.sum(axis=1),axis=0).plot(kind='area')
plt.ylim([0,1])
plt.title('Proportion of Topic over Time')
plt.xlabel('Week')
```

Slight fluctiations, but nothing striking. Perhaps changes were more at the individual level.

```python
type_prop = ctr[ctr['type'].isin(top_types[0:10])].set_index('send_time').groupby([lambda x: x.weekofyear,'type']).count()['content_id'].unstack()
type_prop.divide(type_prop.sum(axis=1),axis=0).plot(kind='area')
plt.ylim([0,1])
plt.title('Proportion of Content-Types over Time')
plt.xlabel('Week')
```

## Variaton in Topic Popularity

Do topics vary in popularity?
How much constistency is there in the popularity of topics among users?

```python
user_topic_ctr = ctr.loc[ctr['topic'].isin(top_topics),['user_id','clicked','not_clicked','topic']].groupby(['user_id','topic']).sum()
user_topic_ctr['total'] = user_topic_ctr.sum(axis=1)
user_topic_ctr['CTR'] = user_topic_ctr['clicked']/(user_topic_ctr['clicked']+user_topic_ctr['not_clicked'])
```

```python
user_topic_ctr['CTR'].unstack().mean().sort_values(ascending=False)
```

```python
sns.boxplot(user_topic_ctr['CTR'].unstack(),orient='h',fliersize=3,whis=[2,98])
plt.xlabel('CTR')
plt.title('Distribution of User CTRs by Topic')
```

Overall popularity of topics varies significantly. On average, some topics are much more popular than others.
Notably, response to different topic types differs by a large amount between users. This is a strong argument for well-personalized emails to users. 

This plot needs to be checked for overly large inequality between the number of links of a given topic which are sent to users. Topics with few questions sent to each user can be noisy. Below we see that there are more than 10 questions for each type, and up to 25. Good enough . . .

```python
a = user_topic_ctr['CTR'].unstack().mean(axis=0)
b = user_topic_ctr['CTR'].unstack().std(axis=0)
c = user_topic_ctr['total'].unstack().sum(axis=0)//1000

CC = np.round(pd.DataFrame({'meanCTR':a,'stdCTR':b,'total (thousands)':c}).sort_values(by='total (thousands)',ascending=False),2)
CC
```

```python
CC.corr()
```

High-Variance Content with Low Volume - should try to target.
High Volume Content with low popularity. Is this because users will read only so many articles on a given topic? Is higher response to the minority topics due to novelty?

Clear negative correlation between number of links of a given topic and average CTR and variation in CTR.
